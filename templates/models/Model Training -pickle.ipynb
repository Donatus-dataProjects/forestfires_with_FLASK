```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('Algerian_forest_fires_cleaned_dataset.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
#Droping unnecessary columns
df.drop(['day', 'month', 'year'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
#Count of unique values
df['Classes'].value_counts()
```

```python
#Feature encoding of the Classes column
df['Classes'] = np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df['Classes'].value_counts()
```

```python
#Converting the 'Classes' column into integer, any of the method works depending on your data

#df['Classes'] = np.where(df['Classes'].str.contains('not fire'),0,1)
#df['Classes'] = df['Classes'].map({'not fire':0,'fire':1}).astype('int')

```

```python
df.tail()
```

```python
#Converting your fetures in to independent and independent 
x = df.drop('FWI', axis = 1)
y = df['FWI']
```

```python
x.head()
```

```python
y
```

```python
#Train Test Split
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size = 0.25, random_state = 42)
```

```python
x.shape, x_train.shape, x_test.shape
```

```python
#Feature selection based on correlation 
x_train.corr()
```

```python
##Checking for multicollitinearity. ie independent feature shpould not be that much correlated based on some threshold values
plt.figure(figsize = (12, 8))
corr  =x_train.corr()
sns.heatmap(corr, annot = True);
```

```python
#Setting a Threshold for to check for multicollinearity
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
# N/B threshold is set by domain expertise or Product Manager
corr_features = correlation(x_train, 0.85)
```

```python
##droping the threshold features when correlation is more than 0.85 
x_train.drop(corr_features, axis = 1, inplace = True)
x_test.drop(corr_features, axis = 1, inplace = True)
```

```python
x_train.shape, x_test.shape
```

## Feature Scaling or Standardization

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled
```

```python
x_test_scaled
```

## Box Plot to understand the Effect of Standard Scaler


```python
plt.subplots(figsize = (18, 5)) 
plt.subplot(1, 2, 1)
sns.boxplot(data = x_train)
plt.title('X_train Before Scaleng')

plt.subplot(1, 2, 2)
sns.boxplot(data = x_train_scaled)
plt.title('X_train After Scaleng')
plt.show()
```

## Linear Regression Model Apploication

```python
#Importing the necessary libraries
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
linreg = LinearRegression()
```

```python
#fitting the x_train_scaled to the ligression

linreg.fit(x_train_scaled, y_train)
```

```python
#x_test_scaled prediction
y_prediction = linreg.predict(x_test_scaled)
```

```python
#Predicting the mean absolute error of y_test
mae = mean_absolute_error(y_test, y_prediction)
```

```python
print('Mean Absolute Error', mae)
print('R2 Score', score)
```

```python
plt.scatter(y_test, y_prediction);
'''This is a very good model as it follows a linear regression relationship'''
```
